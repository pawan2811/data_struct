```python
#1.  Discuss string slicing and provide examples
'''
String slicing is a way to extract a subset of a string in Python by specifying a range of indices. This allows you to create a new string from an existing one. The syntax for slicing is `string[start:end:step]`, where:

- `start` is the index where the slice begins (inclusive).
- `end` is the index where the slice ends (exclusive).
- `step` specifies the step size or the interval between indices.
'''

### Basic Slicinb
#Extracting a Substring:
  
text = "Hello, World!"
substring = text[0:5]  # Extracts from index 0 to 4
print(substring)  # Output: Hello
   

#Extracting from Start to End:**
 
text = "Hello, World!"
substring = text[7:12]  # Extracts from index 7 to 11
print(substring)  # Output: World
   

### Using Ste

text = "Hello, World!"
substring = text[::2]  # Extracts every second character
print(substring)  # Output: Hlo ol!
   


text = "Hello, World!"
reversed_text = text[::-1]  # Extracts characters in reverse order
print(reversed_text)  # Output: !dlroW ,olleH
   


text = "Hello, World!"
substring = text[:5]  # Extracts from start to index 4
print(substring)  # Output: Hello
   

text = "Hello, World!"
substring = text[7:]  # Extracts from index 7 to the end
print(substring)  # Output: World!
   


text = "Hello, World!"
substring = text[-6:-1]  # Extracts from the 6th to last to the 1st to last character
print(substring)  # Output: World
   


text = "Hello, World!"
substring = text[-1:-6:-1]  # Extracts characters in reverse order from the end
print(substring)  # Output: !dlro
   

#Combining Slicing and Other String Methods

text = "Hello, World!"
substring = text[0:5].upper()  # Extract and convert to uppercase
print(substring)  # Output: HELLO
   


text = "Hello, World!"
modified_text = text[:7] + text[7:12].replace("World", "There") + text[12:]
print(modified_text)  # Output: Hello, There!
    

#String slicing is a powerful feature in Python that allows for flexible manipulation of strings with concise and readable syntax.
```

```python
#2.Explain the key features of lists in Python


#Lists in Python are versatile, mutable, and ordered collections that can hold a variety of data types. Here are the key features of lists:

# 1.Ordered Collection**
#Lists maintain the order of elements. The elements in a list have a specific position or index, starting from 0.

fruits = ["apple", "banana", "cherry"]
print(fruits[0])  
  

# 2.Mutable
# Lists can be modified after their creation. You can add, remove, or change elements.

fruits = ["apple", "banana", "cherry"]
fruits[1] = "blueberry"
print(fruits)  
  

#3. **Heterogeneous Elements**
#Lists can contain elements of different data types, including integers, strings, and even other lists.

mixed_list = [1, "apple", 3.14, [2, 4, 6]]
print(mixed_list)  
  

#4. Dynamic Size
# Lists can grow or shrink in size as needed. You can add or remove elements without needing to define the size initially.

numbers = [1, 2, 3]
numbers.append(4)
print(numbers)  
numbers.pop()
print(numbers)  
  

# 5. **Indexing and Slicing**
#Lists support indexing to access individual elements and slicing to access sublists.

fruits = ["apple", "banana", "cherry", "date"]
print(fruits[1])  
print(fruits[1:3])  
  

# 6. **Comprehensions**
#List comprehensions provide a concise way to create lists.

squares = [x**2 for x in range(5)]
print(squares)  
  

### 7. **Built-in Functions and Methods**
#Python provides many built-in functions and methods to manipulate lists.

numbers = [1, 2, 3, 4, 5]
print(len(numbers)) 
print(min(numbers))  
print(max(numbers))  
  


fruits = ["apple", "banana", "cherry"]
fruits.append("date")
print(fruits)  
fruits.remove("banana")
print(fruits) 
fruits.reverse()
print(fruits)  
  

### 8. **Nested Lists**
#Lists can contain other lists, allowing the creation of multi-dimensional data structures like matrices.

matrix = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
print(matrix[1][2]) 
  

# 9. **Iteration**
#You can iterate over the elements of a list using loops.

fruits = ["apple", "banana", "cherry"]
for fruit in fruits:
      print(fruit)
  

### 10. **Copying Lists**
# Lists can be copied using various methods to avoid unintentional modifications.

original = [1, 2, 3]
copy = original[:]
copy.append(4)
print(original)  
print(copy)  
  

#Lists in Python are highly flexible and essential for handling ordered collections of items, providing a wide range of functionalities to manage and manipulate data.
```

```python
#3.Describe how  to access modifes and delete elements in a list with examples

#In Python, accessing, modifying, and deleting elements in a list can be done easily using indexing, slicing, and various list methods. Here are detailed explanations and examples for each operation:

# Accessing Elements
#Accessing by Index:**
#You can access an element by its index using square brackets `[]`.
fruits = ["apple", "banana", "cherry"]
print(fruits[0])  
print(fruits[1])  
    
#Accessing by Negative Index:**
#Negative indices can be used to access elements from the end of the list.

fruits = ["apple", "banana", "cherry"]
print(fruits[-1])  
print(fruits[-2])  
     

#3. **Slicing:**
   #Slicing can be used to access a sublist by specifying a range of indices.
  
fruits = ["apple", "banana", "cherry", "date"]
print(fruits[1:3])  
print(fruits[:2])   
print(fruits[2:])   
     

### Modifying Elements
#1. **Modifying by Index:**
   # You can change the value of an element by assigning a new value to its index.
  
fruits = ["apple", "banana", "cherry"]
fruits[1] = "blueberry"
print(fruits)  # Output: ['apple', 'blueberry', 'cherry']
     

#2. **Modifying a Range of Elements:**
   # You can modify multiple elements at once using slicing.
 
fruits = ["apple", "banana", "cherry", "date"]
fruits[1:3] = ["blueberry", "cranberry"]
print(fruits)  # Output: ['apple', 'blueberry', 'cranberry', 'date']
     

# Deleting Elements
#1. Deleting by Index:
   #You can remove an element by its index using the `del` statement.
   
fruits = ["apple", "banana", "cherry"]
del fruits[1]
print(fruits)  # Output: ['apple', 'cherry']


#2.Deleting a Range of Elements:
   #ou can remove multiple elements at once using slicing with the `del` statement.
fruits = ["apple", "banana", "cherry", "date"]
del fruits[1:3]
print(fruits)  
     

#3. **Removing by Value:**
   # You can remove an element by its value using the `remove()` method. If the value appears multiple times, only the first occurrence is removed.
   
fruits = ["apple", "banana", "cherry", "banana"]
fruits.remove("banana")
print(fruits) 
     

#4. **Removing and Returning the Last Element:**
   #You can remove and return the last element using the `pop()` method without any arguments.
fruits = ["apple", "banana", "cherry"]
last_fruit = fruits.pop()
print(last_fruit)  
print(fruits)     
     

#5. **Removing and Returning by Index:**
   # You can remove and return an element at a specific index using the `pop(index)` method.
  
fruits = ["apple", "banana", "cherry"]
second_fruit = fruits.pop(1)
print(second_fruit)  
print(fruits)       
     

#6. **Clearing All Elements:**
   #- You can remove all elements from a list using the `clear()` method.
fruits = ["apple", "banana", "cherry"]
fruits.clear()
print(fruits)  
     

#These operations provide a comprehensive set of tools for managing the elements in a list, allowing for flexible and dynamic data manipulation.
```

```python
#4. Compare and contrast tuples and lists with examples

#Tuples and lists are both sequence data types in Python that can store collections of items. However, they differ in several key aspects, including mutability, syntax, and usage.

# 1. **Mutability
   #Lists:Mutable, meaning the items in a list can be modified (e.g., items can be added, removed, or changed).
   #Tuples: Immutable, meaning once a tuple is created, its items cannot be modified.

   #Example:
  
print("List example")
my_list = [1, 2, 3]
my_list[0] = 10  # Modifying the first element
print(my_list)   # Output: [10, 2, 3]

print("Tuple example")
my_tuple = (1, 2, 3)
my_tuple[0] = 10  # Attempting to modify a tuple results in a TypeError


# 2. Syntax
   #Lists:** Created using square brackets `[]`.
   #Tuples:** Created using parentheses `()`.

#Example:**
  
my_list = [1, 2, 3]    # List
my_tuple = (1, 2, 3)   # Tuple
   

#3. Performance
   #Lists: Slightly slower than tuples because they are mutable and have more overhead for managing changes.
   #Tuples: Faster than lists due to their immutability, making them more efficient for fixed collections of items.

  
import timeit

list_time = timeit.timeit(stmt="[1, 2, 3, 4, 5]", number=1000000)
tuple_time = timeit.timeit(stmt="(1, 2, 3, 4, 5)", number=1000000)

print(f"List time: {list_time}")
print(f"Tuple time: {tuple_time}")
   



# 4. Methods
   #Lists: Have several built-in methods such as `append()`, `remove()`, `pop()`, `sort()`, etc., for modifying the list.
   #Tuples: Have fewer methods, mainly `count()` and `index()`, since they do not support modification.


my_list = [1, 2, 3]
my_list.append(4)  # List now is [1, 2, 3, 4]

my_tuple = (1, 2, 3)
   # my_tuple.append(4)  # This would raise an AttributeError
   

# 6. Packing and Unpacking
   #Lists and Tuples:** Both support packing and unpacking, allowing the assignment of multiple variables at once.


print("Tuple unpacking")
a, b, c = (1, 2, 3)
print(a, b, c)  # Output: 1 2 3

print("List unpacking")
x, y, z = [4, 5, 6]
print(x, y, z)  # Output: 4 5 6
   

### Summary:
#Lists: Mutable, versatile, slower, and have many methods for modification. Use when you need to change the data.
#Tuples: Immutable, faster, and have fewer methods. Use when the data should remain constant.
```

```python
#5. Describe the key features of sets and provide examples of their use
'''
A set is a built-in data type in Python that stores an unordered collection of unique elements. 
Sets are useful for tasks that involve membership testing, eliminating duplicates, 
               
    and performing mathematical set operations like unions and intersections.
'''
# Key Features of Sets

print("Unordered Collection:")
#Sets do not maintain the order of elements. When you print a set, the items may appear in a different order than they were added.

#Example
  
my_set = {3, 1, 2}
print(my_set) 


print("Unique Elements:")
#Sets automatically remove duplicate values, so all elements in a set are unique.

   
my_set = {1, 2, 2, 3}
print(my_set)  

print("Mutable:")
#Sets themselves are mutable, meaning you can add or remove elements after the set is created.

  
my_set = {1, 2, 3}
my_set.add(4)  # Adding an element
print(my_set)  # Output: {1, 2, 3, 4}
   

print("No Indexing or Slicing")
#Since sets are unordered, you cannot access elements by index or slice them as you would with lists or tuples.
my_set = {1, 2, 3}
   
   

print("Mathematical Set Operations")
#Sets support operations like union, intersection, difference, and symmetric difference.

   
set_a = {1, 2, 3}
set_b = {3, 4, 5}

union_set = set_a | set_b           # Union: {1, 2, 3, 4, 5}
intersection_set = set_a & set_b    # Intersection: {3}
difference_set = set_a - set_b      # Difference: {1, 2}
sym_diff_set = set_a ^ set_b        # Symmetric Difference: {1, 2, 4, 5}

print(union_set)        
print(intersection_set) 
print(difference_set)   
print(sym_diff_set)    
   


print("Mathematical Operations:")
#Use sets to find common elements (intersection), all elements (union), and differences between two collections.


fruits = {"apple", "banana", "cherry"}
tropical = {"banana", "mango", "pineapple"}

common_fruits = fruits & tropical  # Intersection
print(common_fruits)  
```

In Python, dictionary keys must be immutable. This is crucial for ensuring that dictionaries function correctly as data structures. Let's discuss why immutability is important and provide some examples.

Why Must Dictionary Keys Be Immutable?
Hashing Requirement:

Dictionaries in Python are implemented using hash tables. When you add a key-value pair to a dictionary, the key is hashed, and this hash is used to determine where the key-value pair is stored.
If a key were mutable (i.e., its value could change), then its hash value could change as well. This would make it impossible for Python to find the key in the dictionary, as it would no longer be stored at the location determined by its original hash.
Consistency:

Immutability ensures that the keys in a dictionary remain consistent throughout the lifetime of the dictionary. If keys could change, it would lead to unpredictable behavior and could break the integrity of the data structure.
Performance:

Immutable keys guarantee that the hash value remains constant, allowing for fast lookups, inserts, and deletions. Mutable keys would degrade performance since Python would have to rehash the key whenever it was modified, complicating the process and slowing down operations.
Examples of Immutable and Mutable Objects in Python:
Immutable Objects:

Integers: 1, 42, -7
Strings: 'hello', 'Python'
Tuples: (1, 2, 3), ('a', 'b', 'c')
Booleans: True, False
These objects cannot be changed after they are created, making them safe to use as dictionary keys.

Mutable Objects:

Lists: [1, 2, 3], ['a', 'b', 'c']
Dictionaries: {'key': 'value'}, {1: 2, 3: 4}
Sets: {1, 2, 3}, {'a', 'b', 'c'}
These objects can be modified after creation, so they cannot be used as dictionary keys.

```python
# Using strings (immutable) as keys
person = {
    'name': 'Alice',
    'age': 30,
    'city': 'New York'
}

# Using a tuple (immutable) as a key
coordinates = {
    (40.7128, -74.0060): 'New York',
    (34.0522, -118.2437): 'Los Angeles'
}

print(person['name'])  # Outputs: Alice
print(coordinates[(40.7128, -74.0060)])  # Outputs: New York
```
